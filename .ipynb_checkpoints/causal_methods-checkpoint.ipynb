
## Doubly Robust Estimation: A Comprehensive Guide

### Introduction to Doubly Robust Estimation

**Doubly Robust Estimation** is a technique used to estimate causal treatment effects (such as the **Average Treatment Effect (ATE)**) in observational studies where randomization is not feasible. It combines two models to estimate the causal effect of a treatment while controlling for confounding biases and treatment selection biases:

1. **Propensity Score Model**: This model estimates the probability of receiving the treatment based on observed covariates (e.g., it estimates the likelihood that an individual receives treatment given their characteristics).
   
2. **Outcome Model**: This model estimates the potential outcomes (i.e., the counterfactual outcomes) for both treated and untreated groups, based on covariates.

The term "doubly robust" refers to the fact that this method provides an unbiased estimate if **either** the propensity score model or the outcome model is correctly specified, but **not necessarily both**. This is a key advantage over methods that rely on a single model being correctly specified.

### The Doubly Robust Estimator Equation for ATE

The **Doubly Robust Estimator** for the **Average Treatment Effect (ATE)** is given by:

$$
ATE = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{T_i (Y_i - \hat{\mu_1}(X_i))}{\hat{e}(X_i)} + \frac{(1 - T_i)(Y_i - \hat{\mu_0}(X_i))}{1 - \hat{e}(X_i)} + \hat{\mu_1}(X_i) - \hat{\mu_0}(X_i) \right)
$$ 

Where:
- $Y_i$ is the observed outcome for unit $i$.
- $T_i$ is the treatment indicator for unit $i$ (1 if treated, 0 if control).
- $\hat{\mu_1}(X_i)$ is the predicted outcome for unit $i$ if treated (from the outcome model).
- $\hat{\mu_0}(X_i)$ is the predicted outcome for unit $i$ if untreated (from the outcome model).
- $\hat{e}(X_i)$ is the estimated propensity score for unit $i$ (probability of receiving treatment $T_i = 1$ given covariates $X_i$).
- $n$ is the number of units in the dataset.

### Interpretation of Results

- **ATE**: The **Average Treatment Effect** represents the average difference in outcomes between the treatment and control groups.
- **Confidence Interval**: The **95% confidence interval** for the ATE gives you the range of values within which the true ATE is likely to fall, with 95% confidence.
- **Statistical Significance**: This is assessed based on whether the confidence interval excludes zero. If the interval doesn't contain zero, the effect is considered **statistically significant**.

### Why Use Doubly Robust Estimation?

Doubly robust estimation is particularly useful for estimating treatment effects in observational studies where there might be confounding factors. By using both a **propensity score model** and an **outcome model**, the method provides a more reliable estimate of the treatment effect than models relying on a single model. Some key advantages include:

- **Robustness**: The method remains unbiased as long as **one** of the models (either the propensity model or the outcome model) is correctly specified. 
- **Flexibility**: It can handle complex confounding structures and multiple covariates, allowing it to handle more sophisticated datasets.
- **Efficiency**: It tends to be more efficient than simpler methods like matching, especially in large datasets.

### Pros and Cons of Doubly Robust Estimation

| **Pros** | **Cons** |
|----------------------|-----------------------------|
| **Robustness**: The method remains unbiased even if one of the models (propensity score or outcome model) is misspecified. | **Requires Accurate Model Specification**: If both models are misspecified, the estimates may still be biased. |
| **Controls Confounding**: It accounts for confounding variables and helps reduce bias in observational studies. | **Requires More Complex Models**: Estimating both propensity scores and outcome models can be computationally intensive. |
| **Works with Observational Data**: Useful for non-randomized settings where treatment assignment is not random. | **Risk of Overfitting**: Complex models, especially those using machine learning methods, may lead to overfitting. |
| **Improved Efficiency**: More efficient than methods like matching, especially in large datasets. | **Requires Sufficient Common Support**: Both treated and control groups should have overlap in their covariates for reliable estimation. |

### Key Insights from the Doubly Robust Estimator

The beauty of **Doubly Robust Estimation** lies in its **flexibility**. It is **robust** to misspecifications in either the propensity score model or the outcome model. If either one of them is correctly specified, the estimate of the treatment effect remains unbiased.

### How to Apply Doubly Robust Estimation in Python

Here, we'll walk through a practical example using Python. The method combines **propensity score estimation** (via Logistic Regression) and **outcome model estimation** (via Linear Regression), and then uses these models to estimate the **Average Treatment Effect (ATE)**. We also include **bootstrapping** to calculate confidence intervals for the ATE estimate.

#### Functions for Doubly Robust Estimation

1. **`doubly_robust_model`**: This function estimates the **ATE** using doubly robust estimation by fitting both a **propensity score model** and an **outcome model** (for both treated and untreated units). It applies the doubly robust formula to compute the ATE for each observation and returns the average across all units.

2. **`doubly_robust`**: This function performs **bootstrapping** to obtain a **95% confidence interval** for the ATE. Bootstrapping helps assess the precision and statistical significance of the estimated treatment effect.

#### The `evaluate_performance_metrics` Function

In the context of evaluating the performance of the **doubly robust model**, it's important to assess not just the causal estimate but also the **model's predictive performance**. This is where the `evaluate_performance_metrics` function comes into play. It evaluates and interprets key performance metrics like **AUC-ROC** and **Mean Squared Error (MSE)** for both the control and treated groups.

Hereâ€™s what it does:

- **AUC-ROC Evaluation**: The function evaluates the **AUC-ROC** (Area Under the Receiver Operating Characteristic Curve) to measure the quality of the propensity score model. A higher AUC (e.g., greater than 0.80) indicates a better model for predicting treatment assignment. The function categorizes the AUC score into **"Good"** or **"Poor"** based on the threshold.
  
- **MSE Evaluation**: The function evaluates the **Mean Squared Error (MSE)** for both the control and treated groups. The MSE gives us an indication of how well the outcome models (for both treated and control groups) fit the data:
  - **Low MSE (<500)**: Indicates a good fit of the model to the observed data.
  - **Moderate MSE (500-2000)**: Suggests that the model could be improved.
  - **High MSE (>2000)**: Signals poor model fit, requiring further refinement or model adjustments.

The function returns a dictionary with the performance evaluation of both the **AUC-ROC** and **MSE** metrics, providing a comprehensive picture of the model's performance.

### How to Get Started:

1. **Prepare Your Data**: Ensure you have a dataset with a **treatment variable (`T`)**, an **outcome variable (`Y`)**, and relevant **covariates (`X`)**.
   
2. **Run `doubly_robust_model`**: This function will give you a point estimate of the **ATE**, along with important performance metrics (such as **AUC-ROC** and **MSE**).

3. **Run `doubly_robust`**: This function performs **bootstrapping** to provide a **95% confidence interval** for the **ATE**, allowing you to assess the precision of your estimate and its statistical significance.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
def evaluate_performance_metrics(auc_roc, mse_control, mse_treated):
    """
    Helper function to evaluate model performance metrics (AUC-ROC, MSE).
    
    Args:
        auc_roc (float): The AUC-ROC score.
        mse_control (float): The MSE for the control group.
        mse_treated (float): The MSE for the treated group.
        
    Returns:
        dict: Evaluation of performance metrics.
    """
    # Evaluate AUC-ROC
    if auc_roc >= 0.80:
        auc_eval = "Good"
    else:
        auc_eval = "Poor"
    
    # Evaluate MSE for Control and Treated groups
    mse_eval = {}
    mse_statements = {}
    
    for group, mse in zip(["Control", "Treated"], [mse_control, mse_treated]):
        if mse < 500:
            mse_eval[group] = "Low"
            mse_statements[group] = "Low MSE indicates that the outcome models are fitting well to the observed outcomes."
        elif mse < 2000:
            mse_eval[group] = "Moderate"
            mse_statements[group] = "Moderate MSE indicates that the outcome models are fitting not the best but not the worst to the observed outcomes."
        else:
            mse_eval[group] = "High"
            mse_statements[group] = "A High MSE means the outcome model is inaccurate, and you may need to improve it (e.g., using more features, trying a different model, or addressing outliers)."
    
    return {
        'AUC-ROC': auc_eval,
        'MSE Control': mse_eval['Control'],
        'MSE Control Explaination': mse_statements['Control'], 
        'MSE Treated': mse_eval['Treated'],
        'MSE Treated Explaination': mse_statements['Treated']
    }

def doubly_robust_model(df, X, T, Y):
    """
    This function trains a double robust model to estimate ATE.
    
    Args: 
        df (pd.DataFrame): Dataframe with all of the needed features
        X (list): List of feature columns to be used for prediction
        T (string): Treatment column name (binary: 0 or 1)
        Y (string): Outcome column name
    
    Returns:
        dict: ATE, AUC-ROC, MSE for control and treated groups, and performance evaluation.
    """
    
    # Step 1: Propensity score estimation
    ps_model = LogisticRegression(C=1e6, max_iter=1000)
    ps = ps_model.fit(df[X], df[T]).predict_proba(df[X])[:, 1]
    
    # Calculate AUC-ROC for the propensity model
    auc_roc = roc_auc_score(df[T], ps)

    # Step 2: Outcome model for control group (mu0) and treatment group (mu1)
    mu0_model = LinearRegression()
    mu0 = mu0_model.fit(df.loc[df[T] == 0][X], df.loc[df[T] == 0][Y].astype(float)).predict(df[X])
    
    mu1_model = LinearRegression()
    mu1 = mu1_model.fit(df.loc[df[T] == 1][X], df.loc[df[T] == 1][Y].astype(float)).predict(df[X])
    
    # MSE for outcome models
    mse_control = np.mean((df.loc[df[T] == 0][Y] - mu0[df[T] == 0]) ** 2)
    mse_treated = np.mean((df.loc[df[T] == 1][Y] - mu1[df[T] == 1]) ** 2)
    
    # Step 3: Doubly robust estimator for ATE
    term1 = df[T] * (df[Y] - mu1) / ps
    term2 = (1 - df[T]) * (df[Y] - mu0) / (1 - ps)
    
    ate = term1 + mu1 - term2 - mu0
    
    # Get model performance evaluation
    performance_eval = evaluate_performance_metrics(auc_roc, mse_control, mse_treated)
    
    # Return the ATE, performance metrics, and the models for further interpretation
    return {
        'ATE': np.round(ate.mean(), 3),
        'AUC_ROC': np.round(auc_roc, 3),
        'MSE_control': np.round(mse_control, 3),
        'MSE_treated': np.round(mse_treated, 3),
        'Performance Evaluation': performance_eval
    }


def doubly_robust(df, X, treatment_cols, target_cols, boot=250):
    """
    This function sets up DR causal inference. It outputs the impact of treatment on the outcome/target.
    
    Args:
        df (pandas.DataFrame): Dataframe with all features
        X (list): List of X features to be trained
        treatment_cols (list): List of treatment column/s
        target_cols (list): List of Y (outcome) values
        boot (int): Number of bootstrap resamples
    
    Returns:
        pd.DataFrame: A dataframe with treatment, outcome, bootstrapped ATE, lower and upper bounds, and statistical significance
    """
    
    np.random.seed(13)
    plt.rcParams["figure.figsize"] = (6, 4)

    # Initialize a dataframe to store results
    results = []
    
    # Store cumulative AUC and MSE across all iterations
    auc_roc_list = []
    mse_control_list = []
    mse_treated_list = []
    performance_evaluations = []
    
    for Y in target_cols:
        for T in treatment_cols:
            ates = []
            
            # Bootstrap resampling
            for _ in range(boot):
                try:
                    sampled = df.sample(frac=1, replace=True)
                    model_results = doubly_robust_model(sampled, X, T, Y)
                    ates.append(model_results['ATE'])
                    
                    # Collecting AUC, MSE, and performance evaluations for later reporting
                    auc_roc_list.append(model_results['AUC_ROC'])
                    mse_control_list.append(model_results['MSE_control'])
                    mse_treated_list.append(model_results['MSE_treated'])
                    performance_evaluations.append(model_results['Performance Evaluation'])
                    
                except Exception as e:
                    print(f"Error during bootstrap resampling: {e}")
                    pass

            # Calculate average ATE and CI bounds
            ates = np.array(ates)
            lower_bound = np.percentile(ates, 2.5)
            upper_bound = np.percentile(ates, 97.5)
            ate_avg = np.mean(ates)
            
            # Statistical significance check
            significance = "Significant" if lower_bound > 0 or upper_bound < 0 else "Not Significant"
            
            # Add results to the dataframe
            results.append({
                "Treatment": T,
                "Outcome": Y,
                "ATE": np.round(ate_avg, 3),
                "Lower Bound": np.round(lower_bound, 3),
                "Upper Bound": np.round(upper_bound, 3),
                "Statistical Sig": significance
            })
            
            # Print summary ATE and confidence interval for each treatment-outcome pair
            print(f"\nTreatment: {T} on Outcome: {Y}")
            print(f"  Average ATE: {np.round(ate_avg, 3)}")
            print(f"  95% Confidence Interval: ({np.round(lower_bound, 3)}, {np.round(upper_bound, 3)})")
            print(f"  Statistical Significance: {significance}")
            
            # Plot bootstrap distribution
            plt.figure(figsize=(5, 3))
            sns.histplot(ates, kde=False)
            plt.vlines(np.round(lower_bound, 3), 0, 20, linestyles="dotted", label="2.5th Percentile")
            plt.vlines(np.round(upper_bound, 3), 0, 20, linestyles="dotted", label="97.5th Percentile")
            plt.title(f"ATE Bootstrap Distribution for {T} on {Y}")
            plt.legend()
            plt.show()

    # Calculate cumulative performance metrics (average AUC, MSE for control and treated)
    avg_auc_roc = np.round(np.mean(auc_roc_list), 3)
    avg_mse_control = np.round(np.mean(mse_control_list), 3)
    avg_mse_treated = np.round(np.mean(mse_treated_list), 3)

    # Interpret the model performance
    performance_interpretation = evaluate_performance_metrics(avg_auc_roc, avg_mse_control, avg_mse_treated)
    
    # Add cumulative metrics to the dataframe
    results_df = pd.DataFrame(results)
    results_df["Propensity Score"] = avg_auc_roc
    results_df["MSE (Control)"] = avg_mse_control
    results_df["MSE (Treated)"] = avg_mse_treated
    results_df["Propensity Interp"] = [performance_interpretation["AUC-ROC"]] * len(results_df)
    results_df["MSE Control Interp"] = [performance_interpretation["MSE Control"]] * len(results_df)
    results_df["MSE Treated Interp"] = [performance_interpretation["MSE Treated"]] * len(results_df)
    
    # Print cumulative performance metrics
    print("\nCumulative Performance Metrics (across all bootstrap iterations):")
    print(f"  Average AUC-ROC: {avg_auc_roc}")
    print(f"  Average MSE (Control Group): {avg_mse_control}")
    print(f"  Average MSE (Treated Group): {avg_mse_treated}")
    
    print("\nOverall Model Performance Interpretation:")
    print(f"  AUC-ROC: {performance_interpretation['AUC-ROC']}")
    print(f"  MSE (Control Group): {performance_interpretation['MSE Control']}")
    print(f"  Explaination: {performance_interpretation['MSE Control Explaination']}")
    print(f"  MSE (Treated Group): {performance_interpretation['MSE Treated']}")
    print(f"  Explaination: {performance_interpretation['MSE Treated Explaination']}") 

    return results_df
```

```python
# Load Titanic dataset from seaborn
titanic = sns.load_dataset('titanic') 

# Preprocess the dataset: select relevant columns
# 'survived' is the treatment variable, 'fare' is the outcome variable
# We'll use 'age', 'sex', 'class' as the features (X)
X = ['age', 'sex', 'class']
T = 'survived'  # Treatment variable
Y = 'fare'  # Outcome variable 

# Convert categorical columns ('sex' and 'class') to numerical variables
# Convert 'sex' to numeric: 'male' -> 1, 'female' -> 0
titanic['sex'] = titanic['sex'].map({'male': 1, 'female': 0})

# Convert 'pclass' to numeric: 'First' -> 1, 'Second' -> 2, 'Third' -> 3
titanic['class'] = titanic['class'].map({'First': 1, 'Second': 2, 'Third': 3})

titanic = titanic.dropna(subset=[T, Y]+X)

# Test the doubly robust model on the dataset
ate_result = doubly_robust_model(titanic, X, T, Y)
print(f"Estimated ATE from doubly robust model: {ate_result}") 
```

```python
# Run the doubly robust function with bootstrapping
boot_result = doubly_robust(titanic, X, treatment_cols=[T], target_cols=[Y], boot=250)
```
